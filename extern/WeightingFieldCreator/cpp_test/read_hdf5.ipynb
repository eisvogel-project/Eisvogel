```python
import h5py
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
Ez = h5py.File('ez-000020.00.h5', 'r')
Ez_imag = np.flip(Ez["ez.i"][()], axis = 0)
Ez_real = np.flip(Ez["ez.r"][()], axis = 0)
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 1)
ax.imshow(eps)
```
